```python
# Импортируем библиотеку pandas и читаем файл
import pandas as pd

#Совет - используй конструкцию try except:

try:
    # Твой локальный путь к файлу, который открывается у тебя
    games = pd.read_csv('C:\\Users\\elena.kim\\Downloads\\new_games.csv')
except:
    # Путь к файлу который открывается в тренажёре/у ревьювера
    games = pd.read_csv('/datasets/new_games.csv')
```

```python
#Анализируем набор типов и заполненность данных 
games.info()
```

В датафрейме всего 16 956 строк и 11 колонок из них 4 колонки вещественного формата, 7 - формата object. 
При этом EU Sales, JP Sales - формата object, что выглядит некорректным, потому что у продаж должен быть формат float64, но по заданию мы не 
используем эти столбцы, поэтому не будем заниматься их преобразованием, а вот User Score преобразуем, потому что нам нужно категоризировать этот Series, сейчас он формата Object, а должен быть формата Float64 

```python
#Приводим названия столбцов к верхнему регистру и добавляем нижнее подчеркивание вместо пробелов 
games.columns = games.columns.str.upper()
games.columns = games.columns.str.replace(' ', '_')
```

```python
# Выводим первые строки датафрейма на экран
games.head()
```

```python
# Выводим количество пропущенных строк в датафрейме
games.isna().sum()
```

```python
# Подсчитываем процент строк с пропусками
round(games.isna().sum() / len(games) * 100,2)
```

В столбце Year of release <2% незаполненных данных, можно пренебречь

```python
# Приводим тип данных User Score к численному формату 
games['USER_SCORE'] = pd.to_numeric(games['USER_SCORE'], errors ='coerce')
games[['EU_SALES', 'JP_SALES']] = games[['EU_SALES', 'JP_SALES']].apply(pd.to_numeric, errors='coerce')

print(games.dtypes)
```

```python
games.head()
```

```python
#Отфильтровываю тот датафрейм который нужен для анализа, а именно период с 2000г. по 2013г включительно
filtered_games = games[(games['YEAR_OF_RELEASE'] >= 2000) & (games['YEAR_OF_RELEASE'] <= 2013)]
```

```python
#Вывожу количество строк
filtered_games.info()
```

```python
#Избавляемся от дубликатов
games_cleaned = filtered_games.drop_duplicates ()
games_cleaned.info()
```

```python
round(games_cleaned.isna().sum() / len(games_cleaned) * 100,2)
```

49% оценок пользователей и 44% оценок критиков отсутствуют, что весьма много, однако если продумывать стратегию обработки пропусков, не надо забывать, что оценки - это все такие очень субьективная сущность, поэтому заменять их средним, на мой взгляд будет только искажать данные, поэтому оставляю как есть

```python
#Попробуем категоризовать 
games_cleaned = games_cleaned.copy()
games_cleaned['USER_SCORE_GROUP'] = pd.cut(games_cleaned['USER_SCORE'], bins=[0, 3, 8, 10], labels=["Низкая оценка", "Средняя оценка", "Высокая оценка"])
games_cleaned['CRITIC_SCORE_GROUP'] = pd.cut(games_cleaned['CRITIC_SCORE'], bins=[0, 30, 80, 100], labels=["Низкая оценка", "Средняя оценка", "Высокая оценка"])
```

```python
games_cleaned.head(15)
```

```python
grouped_data = games_cleaned.groupby(['USER_SCORE_GROUP', 'CRITIC_SCORE_GROUP'], as_index=False, observed=True)['NAME'].count()
```

```python
print (grouped_data)
```

```python
print (grouped_data['NAME'].sum())
```

```python
grouped_data_user_score = games_cleaned.groupby(['USER_SCORE_GROUP'], as_index=False, observed= True)['NAME'].count()
print (grouped_data_user_score)
print (grouped_data_user_score['NAME'].sum())
```

```python
grouped_data_critic_score = games_cleaned.groupby(['CRITIC_SCORE_GROUP'], as_index=False, observed= True)['NAME'].count()
print (grouped_data_critic_score)
print (grouped_data_critic_score['NAME'].sum())
```

из очищенного датафрейма в 12 829 строк, с обеими оценками оказались 6132 игры. 
При этом отдельно эксперты проставили 7197 оценок, а пользователи 6506, что странно, потому что кажется, что аудитория пользователей должна быть более многочисленной

По 4305 или 70% оценкам, мнения экспертов и пользователей совпали. 
Всего по одной игре зафиксировано "радикальное" различие в оценках между экспертами и пользователями, когда пользователи дали Высокуй оценку, а эксперты Низкую. 
В остальных играх раличие в оценках также есть но не более чем на 1 грейд: в 607 играх пользователи дали среднюю оценку, а эксперты высокую, и в 1094 игре пользователи наоборот дали высокую оценку, а эксперты среднюю.  

```python
grouped_platform = games_cleaned.groupby(['PLATFORM'], as_index=False)['NAME'].count()
sorted_platforms = grouped_platform.sort_values(by='NAME', ascending=False)
sorted_platforms.head(7)

```
