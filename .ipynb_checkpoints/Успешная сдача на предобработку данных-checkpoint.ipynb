````python
#–ü–æ—Å—á–∏—Ç–∞–µ–º —É–¥–µ–ª—å–Ω—ã–π –≤–µ—Å –ø—Ä–æ–¥–∞–∂ –≤ –∫–∞–∂–¥–æ–º —Ä–µ–≥–∏–æ–Ω–µ: 
columns = ['NA_SALES', 'EU_SALES', 'JP_SALES']
for col in columns:
    Total_sales=filtered_games[col].sum()
    filtered_games[col+'_WEIGHT'] = round(filtered_games[col]/Total_sales*100,2)
    
````

````python
#–ü–æ–ø—Ä–æ–±—É–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–æ–≤–∞—Ç—å 

filtered_games['USER_SCORE_GROUP'] = pd.cut(filtered_games['USER_SCORE'], bins=[0, 3, 8, 10.00001], labels=["–ù–∏–∑–∫–∞—è –æ—Ü–µ–Ω–∫–∞", "–°—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞", "–í—ã—Å–æ–∫–∞—è –æ—Ü–µ–Ω–∫–∞"], right='FALSE')
filtered_games['CRITIC_SCORE_GROUP'] = pd.cut(filtered_games['CRITIC_SCORE'], bins=[0, 30, 80, 100.00001], labels=["–ù–∏–∑–∫–∞—è –æ—Ü–µ–Ω–∫–∞", "–°—Ä–µ–¥–Ω—è—è –æ—Ü–µ–Ω–∫–∞", "–í—ã—Å–æ–∫–∞—è –æ—Ü–µ–Ω–∫–∞"], right='FALSE')
````
