```python
# Создаём исходный список с названиями категорий товаров
cat = ['Электроника', 'Одежда', 'Книги', 'Одежда']

# Добавим новую категорию 'Игрушки' в конец исходного списка
cat.append('Игрушки')
print("Добавим новый элемент 'Игрушки' в конец списка методом append()")
print(cat, end='\\n\\n') 

# Добавим новую категорию 'Спорт' на второе место в списке
cat.insert(1, 'Спорт')
print("Добавим новый элемент 'Спорт' на вторую позицию в списке методом insert()")
print(cat, end='\\n\\n')  

# Добавим новые категории из второго списка add_cat
add_cat = ['Кухня', 'Сад']
cat.extend(add_cat)
print('Добавим новые элементы в список из другого списка методом extend()')
print(cat, end='\\n\\n')

# Удалим первое вхождение элемента 'Одежда' в список
cat.remove('Одежда')
print('Удалим первое вхождение определённого элемента методом remove()')
print(cat, end='\\n\\n') 

# Удалим из списка категорию 'Сад' и сохраним её в новой переменной
popped_cat = cat.pop(2)
print("Удалим из списка категорию 'Сад' и сохраним её в новой переменной методом pop()")
print(f'Переменная popped_cat содержит удалённый элемент {popped_cat}')  
print(cat, end='\\n\\n')

# Полностью очистим список
print('Полностью очистим список методом clear()')
cat.clear()
print(cat)
```

```python
def salary_diff (old_salary, new_salary, tax=13):
    '''
    Формула принимает значения старой зарплаты, 
    новой зарплаты и (опционально) налога. Выводит доходы
    со старой и новой зарплатой, старый и новый доходы 
    в неделю и увеличение дохода с новой зарплатой
    '''
    # Шаг 1
    old_profit = old_salary - old_salary / 100 * tax 
    #('Доход со старой зарплатой:', old_profit)

    # Шаг 2
    new_profit = new_salary - new_salary / 100 * tax 
    #print('Доход с новой зарплатой:', new_profit)

    # Шаг 3
    week_old_profit = old_profit / 4
    #print('Старый доход в неделю:', week_old_profit)

    week_new_profit = new_profit / 4 
    #print('Новый доход в неделю:', week_new_profit)

    # Шаг 4
    answer = week_new_profit - week_old_profit
    print('С новой зарплатой доход Павла в неделю увеличится на:', answer)

# Вызовем функцию
salary_diff(1000, 10000)
print (salary_diff)
```

```python
def income_increase(old_salary, new_salary, tax=0.13):
    old_profit = old_salary * (1 - tax)
    new_profit = new_salary * (1 - tax)
    week_old = old_profit / 4
    week_new = new_profit / 4
    increase = week_new - week_old
    return increase  # возвращаем числовое значение для дальнейшего использования

result = income_increase(70, 100)
print("Доход увеличится на", result, "в неделю")
result = income_increase(80,90)
print (result)
```
