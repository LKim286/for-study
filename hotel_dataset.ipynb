```python
# Импортируем библиотеки
import pandas as pd
import matplotlib.pyplot as plt

# Считываем данные
df = pd.read_csv('C:/Users/elena.kim/Downloads/hotel_dataset.csv')

# Преобразовываем тип `object` в `datetime`
df['user check in datetime'] = pd.to_datetime(df['user check in datetime'])

# С помощью акцессора `dt` создаём новый столбец вида YYYY-MM
df['user_check_in_month'] = df['user check in datetime'].dt.to_period('M')

# Группируем данные
df_agg = df.groupby('user_check_in_month').agg({'client id' : 'count', 'price for the night' : 'mean'})
print (df_agg)
```
