```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#  
df = pd.read_csv('C:/Users/elena.kim/Desktop/___/_01_ /AD_HOC/ ALL HANDS_2025/_2025/ .csv', sep=';')

#     
df['TTD, '] = pd.to_numeric(df['TTD, '], errors='coerce')

#    NaN  'TTD, '
df = df.dropna(subset=['TTD, '])

# :  ,  '  ' == ''
# ,          
df_filtered = df[df['  '] != '']

# ,   
print(f"   : {len(df_filtered)}")

#      
plt.figure(figsize=(10, 8))
sns.boxplot(data=df_filtered, x='TTD, ', hue='  ')

# 
plt.xlabel('TTD, ')
plt.title(' TTD    ( "")')
plt.tight_layout()
plt.show()
```
